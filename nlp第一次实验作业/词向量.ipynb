```python
import zhconv
import jieba
import re
import multiprocessing
import logging
from gensim.models import Word2Vec
from gensim.corpora import WikiCorpus
from gensim.test.utils import get_tmpfile
from gensim.models.word2vec import LineSentence
```

```python
# 将语料文件存储为txt文件
def wikiToTxt():
    # This function takes about 25 minutes
    wiki_corpus = WikiCorpus('./zhwiki-latest-pages-articles.xml.bz2', dictionary={})
    texts_num = 0
    train = []
    with open('./wiki_texts.txt', 'w', encoding='utf-8') as output:
        for text in wiki_corpus.get_texts():
            output.write(" ".join(text) + "\n")
            texts_num += 1
            if(texts_num % 100 == 0):
                print(texts_num)
    output.close()
```

```python
# 繁体转换为简体
def convert():
    file = open('./wiki_texts.txt', 'r', encoding = 'utf-8')
    num = 0
    with open('./wiki_simplified.txt', 'w', encoding='utf-8') as output:
        for line in file.readlines():
            num += 1
            output.write(zhconv.convert(line, 'zh-hans'))
            if(num % 100 == 0):
                print(num)
    output.close()
```

```python
# 过滤掉非中文字符
def filter():
    file = open('./wiki_simplified.txt', 'r', encoding='utf-8')
    num = 0
    with open('./wiki_filtered.txt', 'w', encoding='utf-8') as output:
        for line in file.readlines():
            pattern = re.compile(r'[^\u4e00-\u9fa5]')
            chinese = re.sub(pattern, '', line)
            output.write(chinese + '\n')
            num += 1
            if(num % 100 == 0):
                print(num)
    output.close()
```

```python
# jieba分词,并去掉常用停用词
def splite():
    file = open('./wiki_filtered.txt', 'r', encoding='utf-8')
    stoplist = [line.strip() for line in open('./stopwords.txt', 'r', encoding='utf-8').readlines()]
    res = ""
    num = 0
    with open('./wiki_splite.txt', 'w', encoding='utf-8') as output:
        for line in file.readlines():
            line = line.strip("\n")
            words = jieba.cut(line)
            for word in words:
                if word not in stoplist:
                    res += word + " "
            output.write(res + '\n')
            res = ""
            num += 1
            if(num % 100 == 0):
                print(num)
    output.close()
```

```python
# 创建模型
def createModel():
    input = './wiki_splite.txt'
    output = './word2vec1.model'
    logging.basicConfig(format='%(asctime)s:%(levelname)s:%(message)s', level=logging.INFO)
    model = Word2Vec(LineSentence(input), size=400, window=8, min_count=5,
                     workers=multiprocessing.cpu_count())
    model.save(output)
```

```python
def main():
    #wikiToTxt()
    #convert()
    #filter()
    #splite()
    createModel()
    print()

if __name__ == '__main__':
    main()
```
